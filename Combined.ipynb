```python
#Import Dependencies
import pandas as pd
from sqlalchemy import create_engine
```

```python
#Import Super_Bowls CSV
csv_file  = "super_bowls.csv"
super_bowls_df = pd.read_csv(csv_file)
super_bowls_df.head()
```

```python
#Limiting Super_Bowls columns
new_super_bowls_df = super_bowls_df[['Year', 'super_bowl', 'team_winner', 'team_loser', 'attendance','qb_winner_1', 'qb_loser_1', 'coach_winner', 'coach_loser', 'combined_pts', 'difference_pts']].copy()
new_super_bowls_df.head()
```

```python
# Create a filtered dataframe from specific columns
new_super_bowls_df_cols = ["Year", "super_bowl", "team_winner", "team_loser", "attendance", "qb_winner_1", "qb_loser_1", "coach_winner", "coach_loser", "combined_pts", "difference_pts"]
new_super_bowls_df_transformed = new_super_bowls_df[new_super_bowls_df_cols].copy()

# Rename the column headers
new_super_bowls_df_transformed = new_super_bowls_df_transformed.rename(columns={"Year": "year", "qb_winner_1": "qb_winner",
                                                            "qb_loser_1": "qb_loser"})

# Clean the data by dropping duplicates 
new_super_bowls_df_transformed.drop_duplicates("year", inplace=True)

new_super_bowls_df_transformed.head()
```

```python
#Import tv data CSV
tv_file  = "tv.csv"
tv_df = pd.read_csv(tv_file)
tv_df.head()
```

```python
#Limiting tv columns
new_tv_df = tv_df[['super_bowl', 'network']].copy()
new_tv_df.head()
```

```python
# Create a filtered dataframe from specific columns
new_tv_df_cols = ["super_bowl", "network"]
new_tv_df_transformed = new_tv_df[new_tv_df_cols].copy()


# Clean the data by dropping duplicates 
new_tv_df_transformed.drop_duplicates("super_bowl", inplace=True)

new_tv_df_transformed.head()
```

```python
#Import MVP data CSV
mvp_file  = "Name_MVP.csv"
Name_MVP_df = pd.read_csv(mvp_file)
Name_MVP_df.head()
```

```python
#Limiting Super_Bowls columns
new_Name_MVP_df = Name_MVP_df[['Year', 'Super_bowl', 'MVP']].copy()
new_Name_MVP_df.head()
```

```python
# Create a filtered dataframe from specific columns
new_Name_MVP_df_cols = ["Year", "Super_bowl", "MVP"]
new_Name_MVP_df_transformed = new_Name_MVP_df[new_Name_MVP_df_cols].copy()
```

```python
#Connection to postgres
rds_connection_string = "postgres:postgres@localhost:5432/Super_Bowl_Analysis"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
#Name_MVP data to database
new_Name_MVP_df_transformed.to_sql(name='NAME_MVP', con=engine, if_exists='append', index=False)
```

```python
#Super_bowls data to database
new_super_bowls_df_transformed.to_sql(name='SUPER_BOWLS', con=engine, if_exists='append', index=False)
```

```python
#TV data to database
new_tv_df_transformed.to_sql(name='TV', con=engine, if_exists='append', index=False)
```

```python
#Display table names
engine.table_names()
```

```python
#Read Super_bowls data
pd.read_sql_query('select * from "SUPER_BOWLS"', con=engine).head()
```
